```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
from mpld3.display import display_d3
```

```python
# Now make a matplotlib plot

fig, ax = plt.subplots()
x, y = np.random.normal(0, 1, (2, 300))
ax.plot(x[::2], y[::2], 'ro', markersize=10, alpha=0.3)
ax.plot(x[1::2], y[1::2], 'bo', markersize=20, alpha=0.1)

ax.set_xlabel('x axis')
ax.set_ylabel('y axis')
ax.set_title('Matplotlib Plot Rendered in D3!')

display_d3(fig)  # display the output using d3 SVG
```
